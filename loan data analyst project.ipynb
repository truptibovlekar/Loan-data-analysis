```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:/Users/trupti bovlekar/Downloads/Python_projects/Python_projects/Loan_data_analysis_project/Loan_dataset.csv")
df
```

```python
df.columns
```

```python
# 1.What is the average current loan amount for each loan status?
# Group by 'Loan Status' and calculate the mean of 'Current Loan Amount'
avg_loan_by_status = df.groupby('Loan Status')['Current Loan Amount'].mean()
print(avg_loan_by_status)
```

```python
# 2. How does the credit score vary with the annual income?

# Scatter plot: Credit Score vs Annual Income
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Annual Income', y='Credit Score')
plt.title('Credit Score vs Annual Income')
plt.xlabel('Annual Income')
plt.ylabel('Credit Score')
plt.show()
```

```python
# 3. Is there a correlation between the number of open accounts and the current credit balance?

# Scatter plot: Open Accounts vs Current Credit Balance
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Number of Open Accounts', y='Current Credit Balance')
plt.title('Open Accounts vs Current Credit Balance')
plt.xlabel('Number of Open Accounts')
plt.ylabel('Current Credit Balance')
plt.show()
```

```python
# 4. What is the distribution of credit scores across different home ownership types?

# Box plot: Credit Score by Home Ownership
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Home Ownership', y='Credit Score')
plt.title('Credit Score by Home Ownership')
plt.xlabel('Home Ownership')
plt.ylabel('Credit Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. How does the annual income differ for different purposes of loans?

# Box plot: Annual Income by Purpose
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Purpose', y='Annual Income')
plt.title('Annual Income by Loan Purpose')
plt.xlabel('Purpose')
plt.ylabel('Annual Income')
plt.xticks(rotation=45)
plt.show()
```
